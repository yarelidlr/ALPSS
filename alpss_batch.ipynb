```python
from alpss.alpss_main import alpss_main
import json
from pathlib import Path
```

```python
folder = Path( '/Users/alirachidi/dev/work/misc_scripts/alpss/bad_oscilloscope_signals')
files = [str(f) for f in folder.iterdir() if f.is_file() and f.suffix == ".csv"]
print(files[0])
```

```python

with open('/Users/alirachidi/dev/work/helix/alpss-dagster/alpss_config.json', 'r') as f:
    config = json.load(f)

config['io']['out_files_dir'] = '/Users/alirachidi/dev/work/misc_scripts/alpss/bad_oscilloscope_signals/output'

for file in files:
    if "JHAMAL00018" in file:
        config['material']['density'] = 4428
        config['material']['C0'] = 5130
        config['material']['C_L'] = 6130
    elif "JHAMAL00016" in file:
        config['material']['density'] = 4507
        config['material']['C0'] = 4726
        config['material']['C_L'] = 6070
    print(f"Processing file {file}...")
    config['io']['filepath'] = file
    
    try:
        fig, items = alpss_main(**config)
    except Exception as e:
        print(f"Error occurred while processing {file}: {e}")
        continue
```
