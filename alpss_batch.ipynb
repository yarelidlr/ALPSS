```python
from alpss.alpss_main import alpss_main
import json
from pathlib import Path
import pandas as pd
```

```python
folder = Path( '/Users/alirachidi/dev/work/misc_scripts/alpss/bad_oscilloscope_signals')
files = [str(f) for f in folder.iterdir() if f.is_file() and f.suffix == ".csv"]
print(files[0])
```

```python
with open('/Users/alirachidi/dev/work/helix/alpss-dagster/alpss_config.json', 'r') as f:
    config = json.load(f)

config["io"]['out_files_dir'] = folder / "output"
config["spall"]['spall_enabled'] = True
config["hel"]['hel_enabled'] = True

results_list = []
for file in files:
    if 'JHAMAL00016' in file:
        config['spall']['density'] =  4507
        config['spall']['C0'] =  4726
        config['spall']['C_L'] =  6070
    elif 'JHAMAL00018' in file:
        config['spall']['density'] =  4428
        config['spall']['C0'] =  5130
        config['spall']['C_L'] =  6130
    else:
        continue

    print(f"Processing file {file}...")
    config["io"]['filepath'] = file
    try:
        fig, items = alpss_main(**config)
        result_dict = items['results'][0]
        result_dict['filename'] = Path(file).name
        results_list.append(result_dict)
    except Exception as e:
        print(f"Error processing file {file}: {e}")
        continue
```

```python
# Package results into dataframe
results_df = pd.DataFrame(results_list)
results_df
```

```python
compressed = results_df[['filename', 'Velocity at Max Compression Freq Uncertainty', 'Velocity at Max Compression Vel Uncertainty']].sort_values('Velocity at Max Compression Freq Uncertainty', ascending=False)
compressed.to_csv(folder / "compressed_results.csv", index=False)
```
